```python
df = pd.read_csv('Datasetv2.csv', index_col=0).reset_index()
```

```python
df_from = pd.DataFrame(columns=new_columns, index=df.index)
df_to = pd.DataFrame(columns=new_columns, index=df.index)
```

```python
df_from['Departure City'] = df['Departure City']
df_to['Trip Name'] = df['Trip Name']
```

```python
%%time 

def matching_trip_name(row):
    try:
        trip_name = row['Trip Name']

        coords = trip_name_coord_dict[trip_name]
        row.update(matching_to[coords])
    except:
        pass
    return row

df_to_applied = df_to.apply(matching_trip_name, axis=1)
```

```python
%%time 

def matching_departure_city(row):
    try:
        trip_name = row['Departure City']

        coords = departure_cities_coord_dict[trip_name]
        row.update(matching_from[coords])
    except:
        pass
    return row

df_from_applied = df_from.apply(matching_departure_city, axis=1)
```

```python
add_df = df_to_applied.join(df_from_applied, lsuffix='_to', rsuffix='_from')
```

```python
clear_add_df = add_df.drop(columns=['Trip Name', 'Departure City'])
```

```python
res_df = df.join(clear_add_df)
```

```python
res_df.to_csv('Datasetv2+.csv')
```
